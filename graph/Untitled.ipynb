```julia
using MetaGraphs, LightGraphs, GraphRecipes, Plots

n = 10
g = MetaGraph(n)
g = MetaDiGraph(n, 3.0)

for s = 1:n
    for d = 1:n
        if rand() > 0.5
            add_edge!(g, s, d)
        end
    end
end

function get_width(s, d)
    if (s > 5) & (d > 5)
        return 3.0
    else
        return 1.0
    end
end

for s = 1:n
    for d = 1:n
        set_prop!(g, Edge(s,d), :weight, get_width(s,d))
    end
end

#pyplot()
#plots()
gr()
graphplot(g, edge_width=(s,d,w) -> get_prop(g, Edge(s,d), :weight), 
          arrow=arrow(:closed, :head, 0.1, 0.1),
          markersize = 0.2,
          node_weights = 1:n,
          #markercolor = range(colorant"yellow", stop=colorant"red", length=n),
          markercolor = :white,
          names = 1:n,
          fontsize = 10,
          nodeshape = :circle,
          linecolor = :darkgrey,
          dpi = 200
          #shorten=0.95
          )
```
